```python
import pandas as pd

import matplotlib.pyplot as plt


products = {
    "Products": ["iPhone14", "iPhone15", "iPhone16", "iPhone17"],
    "Sales": [23, 56, 64, 21]
}

product_df = pd.DataFrame(products)

print(product_df)

plt.pie(
    product_df["Sales"],
    labels=product_df["Products"],
    autopct="%1.1f"

)
plt.title("Product Pie")

plt.show()
```
