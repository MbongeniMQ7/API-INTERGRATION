```python
import pandas as pd
import matplotlib.pyplot as plt

department_data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "IT": [10000, 13000, 18000, 19000, 22000, 20000],
    "HR": [12000, 15000, 14000, 1800, 27000, 21000]




}
dept_df = pd.DataFrame(department_data)

print(dept_df)

plt.plot(
    dept_df["Month"],
    dept_df["IT"],
    marker = "o",
    label="IT"
)

plt.plot(
    dept_df["Month"],
    dept_df["HR"],
    marker = "o",
    label="HR"
)

plt.title("MontDepartment  Sales")

plt.xlabel("Month")
plt.ylabel("Sales")


plt.legend()

plt.show()



```
